```python
import shutil
import sys
import os
from pathlib import Path
import json
import pysubs2
from pysubs2 import SSAFile, SSAEvent, make_time
import re
```

```python
root_path = Path('/home/jianghui/dataset/MovieNet-Subset/')

tt_list = []
with open(Path(root_path, 'movie_id_list'), 'r') as f_tt_list:
    for tt_line in f_tt_list:
        tt_list.append(tt_line.replace('\n', ''))
tt_list
```

1. check /img shot sum with /shot sum

```python
%env PYTHONOPTIMIZE = 2
```

```python
for tt_id in tt_list:#['tt0448157']:#
    # img shot sum
    img_path = Path(root_path, 'all/img', tt_id)
    img_shot_sum = sum(1 for _ in img_path.iterdir())
    if __debug__: print(f'img_shot_sum is {img_shot_sum}', end=" ")

    # shot sum
    shot_file = Path(root_path, 'all/shot/', f'{tt_id}.json')
    with open(shot_file) as f_json:
        shot_json = json.load(f_json)
    shot_sum = shot_json[-1]['shot_id'] + 1
    if __debug__: print(f'shot_sum is {shot_sum}', end=' ')

    # scene sum
    scene_file = Path(root_path, 'all/scene/', f'{tt_id}.json')
    with open(scene_file) as f_json:
        scene_json = json.load(f_json)
    scene_shot_sum = scene_json[-1]['shot_end_id'] + 1
    if __debug__: print(f'scene_shot_sum is {shot_sum}')

    assert img_shot_sum == shot_sum, f'{tt_id}'

print("Everything is under control")
```
